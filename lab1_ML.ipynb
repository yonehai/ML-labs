```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
list = np.random.randint(0,100,size=(1000, 2))
df = pd.DataFrame(list, columns=['X', 'Y'])
```

```python
km = KMeans(n_clusters=3).fit(df)
centr = km.cluster_centers_
labels = km.labels_
df['Cluster'] = labels
plt.scatter(df.X, df.Y, c= labels.astype(float), s=20,
alpha=0.5)
plt.scatter(centr[:, 0], centr[:, 1], c='red', s=50)
plt.show()
```

```python
clusters_group = df.groupby('Cluster')
print(clusters_group.describe().T)
```
